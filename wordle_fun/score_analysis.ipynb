```python
import pandas as pd

scores = pd.read_csv("scores.csv", names=["Word", "Prunes"], dtype={"Word": str,"Prunes":str})

```

```python
scores
```

```python
scores["Prunes"] = scores["Prunes"].apply(lambda x: list(map(int, x.replace("[","").replace("]","").split(",")  )))
```

```python
scores
```

```python
scores["Sum"] = scores["Prunes"].apply(lambda x: sum(x))
```

```python
scores.sort_values("Sum")
```

```python
scores["Mean"] = scores["Prunes"].apply(lambda x: sum(x)/len(x))
```

```python
scores.sort_values("Mean")
```

```python
scores["Min"] = scores["Prunes"].apply(lambda x: min(x))
scores["Max"] = scores["Prunes"].apply(lambda x: max(x))
```

```python
scores.sort_values("Min")
```

```python
import statistics
scores["Median"] = scores["Prunes"].apply(lambda x: statistics.median(x))
scores.sort_values("Median")
```
